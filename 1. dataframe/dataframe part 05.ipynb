## part1. membuat dataframe dari list dan modifikasi nilai

sebelum membuat dataframe, pastikan dulu untuk memanggil library pandasnya yak.. untuk membuat dataframe, formatnya cukup mudah

```python
import pandas as pd
```

data yang dibutuhkan untuk kolom dataframe dapat berupa format data list

```python
'''
df = pd.DataFrame({'nama_kolom_1' : datanya
                   'nama_kolom_2' : datanya
                   'nama_kolom_n' : datanya})
'''   

# contoh
df = pd.DataFrame({'nama_siswa' : ['andi','citra','baayu','jack','gilbert',
                                   'stefa','rei','baim','intan','melati'],
                   'mata_kuliah': ['statistika']*10,
                   'nilai' : [90,80,60,68,84,
                              86,98,100,85,98]})
# menampilkan dataframe
# cara yang paling enak menampilkan dataframe adalah dengan menggunakan fungsi display,
# walaupun fungsi print dapat juga digunakan
display(df)
```

untuk dataframe pada python, index barisnya dimulai dari 0, jadi untuk keseluruhan barisnya adalah **n + 1**, apabila pada baris akhirnya menunjukkan angka 9, maka total barisnya adalah 10

misalkan, ternyata nilai jack salah input, seharusnya nilai dia adalah 82, maka untuk mengganti nilainya dapat dilakukan dengan method .replace(), yaitu :

```python
df['nilai'] = df['nilai'].replace(['68'],'82')

display(df)
```

method lainnya pada dataframe dapat dilihat pada link [berikut](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html)

## part2. membuat dataframe dari dictionary dan modifikasi nilai

bisa juga dataframe yang dibuat berasal dari format dictionary, contohnya adalah sebagai berikut :

```python
dictionary = {'nama_siswa' : ['andi','citra','baayu','jack','gilbert',
                                   'stefa','rei','baim','intan','melati'],
              'mata_kuliah': ['statistika']*10,
              'nilai' : [90,90,60,68,84,86,98,100,85,98]}

df2 = pd.DataFrame(dictionary, columns = ['nama_siswa', 'mata_kuliah', 'nilai'])
display(df2)
```

kalau dari format dictionary, yang perlu diingat adalah, **nama kolom di dataframe harus sama dengan key yang ada pada dictionary**

kemudian, misalkan kita lihat bahwa nilai andi dan citra sama nih, salah satu dari mereka merasa ngga terima dan mengajukan revisi nilai, setelah revisi, nilai andi adalah 88 dan nilai citra adalah 92

cara yang paling mudah untuk mengganti suatu nilai pada dataframe adalah dengan menggunakan iloc dan loc, walaupun ada si beberapa cara yang lain..

```python
# bedanya, iloc dan loc adalah sebagai berikut :
''' 
loc[nama baris, nama kolom]
iloc[posisi baris, posisi kolom]
'''

df2.loc[0, 'nilai'] = 88
df2.loc[1, 'nilai'] = 92

display(df2)
```

sebenernya, loc dan iloc termasuk dalam materi indexing dataframe. untuk materi lainnya dapat diakses [disini](https://github.com/mrifqia67/Course-Data-Science/tree/main/2.%20Pandas%20dan%20Data)

## part4. filtering dan labelling dataset

kemudian, apabila kita ingin memfilter suatu data berdasarkan kriteria tertentu, maka dapat dilakukan sebagai berikut :

```python
df3 = pd.DataFrame({'nama' : ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o'],
                    'matkul' : ['statistika']*5 + ['matematika']*5 + ['fisika']*5,
                    'nilai' : [92,87,85,86,98,84,82,80,97,70,85,64,65,85,74]})

display(df3)
```

misalkan ingin memfilter nilai berdasarkan kriteria mata kuliah fisika, maka :

```python
# filter matkul fisika
display(df3[df3['matkul'] == 'fisika'])
```

bisa juga nih filternya berdasarkan beberapa kriteria, misal filter matkul fisika dan nilainya 85, maka :

```python
# filter matkul fisika dan nilai 85
display(df3[(df3['matkul'] == 'fisika') & (df3['nilai'] == 85)])
```

kemudian kita ingin melabelkan data tertentu, misal nilai sekian masuk ke dalam kategori A, B, dan seterusnya. misal nilai A apabila nilainya berada pada rentang 80-100, B berada pada rentang 70-79, C berada pada rentang 50-69, D berada pada rentang 0-49

untuk mengerjakan hal ini, saya lebih suka menggunakan if dalam bentuk list, walaupun terdapat cara menggunakan looping. untuk pengerjaan yang list adalah sebagai berikut :

```python
# membuat kriteria nilai
kriteria_nilai = ['A' if 80 <= x < 100 else 
                  'B' if 70 <= x <  79 else
                  'C' if 50 <= x <  69 else 
                  'D' for x in df3['nilai']]

df3['kriteria_nilai'] = kriteria_nilai
display(df3)
```
